```python
from bayes_opt import BayesianOptimization

import matplotlib.pyplot as plt

import math

import numpy as np
```

```python
# 先ずは、適当な関数を定義
def func(x,a,b,c):
  
    d=900
    #y = a * math.sin( b * math.sin( c * x * math.pi / 180 ) + x * math.pi / 180 )
    #y = a * math.sin( x * math.pi / d ) + b * math.sin( c * x * math.pi / d) 
    #y = math.sin( x * 0.05 + 2)
    #y = math.pow(x/1000-a,2) + math.pow(x/1000-b,1)-c
    #y = math.pow((x-1000)/1000-a,2) + math.pow((x-1000)/1000-b,1)+c
    #y = math.exp((x-1000)/1000*a) + math.sin((x-1000)/5*b*math.pi/180) + math.cos((x-1000)/5*c*math.pi/180)
    y = math.exp((x-1000)/1000*a) + c*math.sin((x-1000)/5*b*math.pi/180)

    return y
```

```python
#size = 500

def plot(a,b,c,a1,b1,c1):
    global imgcount

    size = 2000
    data = np.zeros((size,2))

    plt.clf()

    for i in range(size):
        data[i,0] = i
        data[i,1] = func(i,a,b,c)

    plt.plot(data[:,0], data[:,1])
    
    for i in range(size):
        data[i,0] = i
        data[i,1] = func(i,a1,b1,c1)

    plt.plot(data[:,0], data[:,1])
    plt.savefig("img{0:05d}.png".format(imgcount))    
    plt.show()
   
    imgcount += 1
```

```python
def get_diff(a,b,c):
    a1 = 1
    b1 = 2
    c1 = 0.5
    diff = 0
    for x in range(2000):
        diff += math.pow(func(x,a1,b1,c1) -func(x,a,b,c),2)
    rmse = math.sqrt(diff / 2000)
    
    retval = -math.log(rmse)
    
    plot(a,b,c,a1,b1,c1)
    
    return retval
```

```python
imgcount = 0

pbounds ={
    'a': (-0.0,4.0),
    'b': (-0.0,4.0),
    'c': (-0.0,4.0)
}

bo = BayesianOptimization(f=get_diff, pbounds=pbounds)

bo.maximize(init_points=10, n_iter=100)
```
